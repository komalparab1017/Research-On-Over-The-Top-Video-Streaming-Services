```python
#MIN-MAX SCALER
min_max_scaler = preprocessing.MinMaxScaler(feature_range =(0, 1)) 
new_x= min_max_scaler.fit_transform(data) 
feature_scaled=pd.DataFrame(new_x,columns=data.columns)
```

```python
feature_scaled
```

# Using the elbow method to find the optimal number of clusters

```python

from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(feature_scaled)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss,'go--', color='green')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.grid()
plt.show()
```

from above graph we can see that optimum no of cluster size is 3

# Training the K-Means model on the dataset 

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(feature_scaled)
y_predicted
```

```python
kmeans.cluster_centers_
```

```python
#adding cluster column to data
feature_scaled['cluster']=y_predicted
```

```python
#to check no of observations in each cluster.
feature_scaled['cluster'].value_counts()
```
